```python
def plot_colorline(bSCSCfront, bSTDfront):
    aux = []
    top = max(bSCSCfront)+1
    for i in range(len(bSCSCfront)):
        if (bSTDfront[i]==0):
            aux.append(top)
        else:
            aux.append(bSCSCfront[i]/bSTDfront[i])
    return np.array([aux])
```
